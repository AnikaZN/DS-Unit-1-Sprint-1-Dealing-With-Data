```
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
from bokeh.io import output_file, show
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis6
output_notebook()

N = 4000 #number of dots
x = np.random.random(size=N) * 100 #generate x-value as random point from 0-1, multiply by 100 so graph is full
y = np.random.random(size=N) * 100 #ditto but y-values

radii = np.random.random(size=N) * 1.5 #randomly set size of dots

#colors = ["#%02x%02x%02x", (zip(50+2*x, 30+2*y))] #this will eventually be code that lets me use different colors
colors = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)] #first bit assigns hexcode values, then ???, then the zip() function creates the pattern

output_file('color_scatter.html', title='Color Scatter', mode='cdn') #image created is a CDN 'color_scatter.html' file with the tite of 'Color Scatter'

TOOLS = 'crosshair, pan, wheel_zoom, box_zoom, reset, box_select, lasso_select' #Bokeh-specific tools which appear on right-hand side of image

p = figure(tools=TOOLS, x_range=(0,100), y_range=(0,100)) #sets up x and y range and places tools

p.circle(x,y,radius=radii, fill_color=colors, fill_alpha=.6, line_color=None) #brings all the data together to create the circles in the image, fill_alpha sets opacit and line_color is an outline

show(p)

#I spent about an hour trying to figure out how to change the colors in this
#thing and haven't figured it out yet but I will keep trying
#I also made sure to retype all the code and not just copy-paste
```

```
import matplotlib.pyplot as plt

from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
from bokeh.io import output_file, show
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis6
output_notebook()

df = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/titanic.csv', index_col=0)
features = ['fare', 'pclass']
df.dropna(subset=features, inplace=True)

df.head()
```

```
plt.plot([df.pclass], [df.fare], 'bo')

output_file('titanic.html', title='Rainbow Titanic', mode='cdn')

TOOLS = 'crosshair, pan, wheel_zoom, box_zoom, reset, box_select, lasso_select'

p = figure(tools=TOOLS, x_range=(0,100), y_range=(0,100))

p.circle(x,y,radius=3, fill_color=colors, fill_alpha=2, line_color='white')

plt.axis([0, 4, 0, 200])

plt.show;
```

```
from bokeh.plotting import figure, output_file, show

x = df.pclass
y = df.fare

output_file('titanic.html')

TOOLS = 'crosshair, pan, wheel_zoom, box_zoom, reset, box_select, lasso_select'
p = figure(tools=TOOLS, x_range=(0,4), y_range=(0,200))

p.circle(x, y, size=10)

show(p)
```
